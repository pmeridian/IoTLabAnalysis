```python
print "Reading data from ThingsSpeak with Json"

import urllib, json
url = 'https://thingspeak.com/channels/491957/feed.json?days=7'
response = urllib.urlopen(url)
data = json.loads(response.read())

import pandas as pd
df = pd.DataFrame.from_dict(data['feeds'])
df.head(5)
```

```python
df['time']=pd.to_datetime(df['created_at'])
df['temp'] = df['field4'].astype(float)
df['battery'] = df['field7'].astype(float)
df['charge current'] = df['field8'].astype(float)
#df.info()

import matplotlib.pyplot as plt
plt.figure(); df.plot(x='time',y='temp');
plt.figure(); df.plot(x='time',y='battery');
plt.figure(); df.plot(x='time',y='charge current');

```
